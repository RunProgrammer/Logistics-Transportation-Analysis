Analysis of geoLocation

```python
import folium
from folium.plugins import MarkerCluster

sampleDF = df.sample(n=1000,random_state=42)
#Taking the mean of the latitude and longitude for making it into the map
map_ = folium.Map(location=[sampleDF['vehicle_gps_latitude'].mean(),sampleDF['vehicle_gps_longitude'].mean()],zoom_start=5)
marker_cluster = MarkerCluster().add_to(map_)

for index,row in sampleDF.iterrows():
    folium.Marker(
        location=[row['vehicle_gps_latitude'],row['vehicle_gps_longitude']],
        popup=f"Risk : {row['route_risk_level']}, Congestion : {row['traffic_congestion_level']}",
        icon=folium.Icon(color='red' if row['route_risk_level'] > 6 else 'green')
    ).add_to(marker_cluster)

map_.save("vechileGeoAnalysis.html")
```

Heat map for traffic

```python
from folium.plugins import HeatMap

# map_ = folium.Map(location=[sampleDF['vehicle_gps_latitude'].mean(),sampleDF['vehicle_gps_longitude'].mean()],zoom_start=5)
head_data = [[row['vehicle_gps_latitude'],row['vehicle_gps_longitude'],row['traffic_congestion_level']]
             for index,row in sampleDF.iterrows()]
HeatMap(data=head_data,radius=20,blur=30,max_zoom=1).add_to(map_)

map_.save('heatMap.html')
```

# To see how warehouse affecting logistics/transport

```python
# Bin into categories (you can choose how many bins)
import seaborn as sns

sampleDF['fulfillment_category'] = pd.qcut(
    sampleDF['order_fulfillment_status'], 
    q=5, 
    labels=["Very Low", "Low", "Medium", "High", "Very High"]
)

plt.figure(figsize=(12, 8))
sns.histplot(
    data=sampleDF,
    x='warehouse_inventory_level',
    hue='fulfillment_category',
    multiple='fill',
    stat='probability',
    bins=20,
    palette='mako'
)
plt.title("Warehouse Inventory vs Order Fulfillment Status (Binned)")
plt.ylabel("Proportion")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
